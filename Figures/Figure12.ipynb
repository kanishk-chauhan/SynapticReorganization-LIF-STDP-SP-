```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pylab as pl
import matplotlib.gridspec as gridspec
import scipy.ndimage as ndimage
from matplotlib.ticker import AutoMinorLocator, FormatStrFormatter
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.interpolate import interp1d

mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['axes.linewidth'] = 2

mpl.rcParams['font.weight'] = 'bold'
mpl.rcParams['axes.labelweight'] = 'bold'

plt.rc('text', usetex=True) ## this line is necessary to use the correct fonts 
plt.rc('font', family='serif') # this line is necessary to use the correct font


# This function makes the plots pretty 
def plot_prop(f1,ax):
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_ticks_position('left') 
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['left'].set_position(('axes', -0.01))
    ax.spines['bottom'].set_position(('axes', -0.02))
    ax.xaxis.set_minor_locator(AutoMinorLocator(1))
    ax.yaxis.set_minor_locator(AutoMinorLocator(1))

    # x and y tickers
    plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%g'))
    plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%g'))
    plt.tick_params(direction='out',which='major', length=8,width=1.5,colors='k',labelsize=f1)
    plt.tick_params(direction='out',which='minor',length=4,width=1, color='k')
    return;
```

```python
from numpy import std, mean
N = 400 # number of neurons

Pw_arr = [0,0.001,0.01,0.1,1] 
nSPits = 999

l1=2; fn1=14; fn2=16; fn3=18  # the font sizes 
fig = plt.figure(constrained_layout=True)
fig.set_size_inches(10,5.5)  # the size of the plot in inches 
gs = gridspec.GridSpec(2, 2, figure=fig)  # the gridspace, here I set 2 * 1

ax1 = pl.subplot(gs[0])
plot_prop(fn1,ax1)
ax2 = pl.subplot(gs[1])
plot_prop(fn1,ax2)
ax3 = pl.subplot(gs[2])
plot_prop(fn1,ax3)
ax4 = pl.subplot(gs[3])
plot_prop(fn1,ax4)

ax1.set_xlim(1,1e3)
ax1.set_ylim(4,5.)
ax1.set_xscale('log')
ax1.set_ylabel("Firing rate [Hz]",fontsize=fn3)
ax1.text(-0.23,0.95,'(a)',fontsize=fn3,transform=ax1.transAxes)

# ax2.set_ylim(0.6,1)
ax2.set_xlim(1,1e3)
ax2.set_xscale('log')
# ax2.set_yscale('log')
ax2.set_ylabel("Order parameter",fontsize=fn3)
ax2.text(-0.23,0.97,'(b)',fontsize=fn3,transform=ax2.transAxes)

# ax3.set_ylim(0.05,0.12)
ax3.set_xlim(1,1e3)
ax3.set_xscale('log')
ax3.set_ylabel(r"$\beta(t)$",fontsize=fn3)
ax3.set_xlabel("SP iteration",fontsize=fn3)
ax3.text(-0.23,0.95,'(c)',fontsize=fn3,transform=ax3.transAxes)

# ax4.set_ylim(0.05,0.12)
ax4.set_xlim(1,1e3)
ax4.set_xscale('log')
# ax4.set_yscale('log')
ax4.set_ylabel("CV",fontsize=fn3)
ax4.set_xlabel("SP iteration",fontsize=fn3)
ax4.text(-0.23,0.95,'(d)',fontsize=fn3,transform=ax4.transAxes)
    
colors = ['teal','darkorange','green','red','purple','dimgray','cyan','tomato']

for i in range(len(Pw_arr)):
    Pw = Pw_arr[i]
    data = np.loadtxt("Data/Data_Fig12/STDP+rSP_Pw=%.3f_AverageData.dat"%(Pw))
    odpr_av = data[0]
    fr_av = data[1]
    ndd_av = data[2]
    cv_av = data[3]
    odpr_sd = data[4]
    fr_sd = data[5]
    ndd_sd = data[6]
    cv_sd = data[7]

    ax1.plot(np.arange(0,nSPits),fr_av,lw=l1,color=colors[i])
    ax1.fill_between(np.arange(0,nSPits),y1=fr_av+fr_sd,y2=fr_av-fr_sd,color=colors[i],alpha=0.3)
    ax2.plot(np.arange(0,nSPits),odpr_av,lw=l1,color=colors[i],label=r'$P_\mathtt{w}$=' f'{Pw}')
    ax2.fill_between(np.arange(0,nSPits),y1=odpr_av+odpr_sd,y2=odpr_av-odpr_sd,color=colors[i],alpha=0.3)
    ax3.plot(np.arange(0,nSPits),ndd_av,lw=l1,color=colors[i])
    ax3.fill_between(np.arange(0,nSPits),y1=ndd_av+ndd_sd,y2=ndd_av-ndd_sd,color=colors[i],alpha=0.3)
    ax4.plot(np.arange(0,nSPits),cv_av,lw=l1,color=colors[i])
    ax4.fill_between(np.arange(0,nSPits),y1=cv_av+cv_sd,y2=cv_av-cv_sd,color=colors[i],alpha=0.3)
    
ax2.legend(fontsize=fn2,loc='lower right',frameon=False,handlelength=1.2,handletextpad=0.5)

# plt.savefig('SP-steady_Pw.jpg',dpi=600)
plt.show()
```
